# Predicting Reliance Stock Value using the Volume and Date

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Checking the distribution of data

```python

sns.scatterplot(x="Volume", y="Close", data=data)
```

## Searching for any outliers in the dataset

```python
#detecting outliers Function

def detect_outliers(data):
    outlier=[]
    threshold=3
    mean=np.mean(data)
    std=np.std(data)
    for i in data:
        z_score=(i-mean)/std
        if np.abs(z_score)>threshold:
            outlier.append(i)
    return outlier
```

```python
# Marking the Outliers
x=detect_outliers(data["Volume"])
np.min(x)
```

```python
df1 = data[data["Volume"]>46465281.0]
df1
```

```python
data = pd.concat([data, df1]).drop_duplicates(keep = False)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
scaler = StandardScaler()
```

```python
df = data[["Date", "Close", "Volume"]]
df
df['Date'] = pd.to_datetime(df.Date,format='%Y-%m-%d')
df
```

```python
import datetime as dt

df['Date']=(df['Date'].map(dt.datetime.toordinal))
for date in df['Date']:
    date = int(date)
```

## Standardising Values for training the model

```python
df_new = pd.DataFrame(scaler.fit_transform(df[["Close", "Volume"]]), df["Date"], columns = ['Close', 'Volume'])
df_new["Date"] = df_new.index
df_new.head()
```

```python
x = df_new[["Date", "Volume"]]
y = df_new["Close"]

y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

# Using KNN Algorithm

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn = KNeighborsRegressor(n_neighbors = 4)
knn.fit(x_train, y_train)
```

```python
pred = knn.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

### Finding out the accuracy of the model

```python
r2_score(y_test, pred)
```

```python
pred
```

## Checking for the best value of k

```python
error_rate=[]
for i in range(1, 30):
    knn = KNeighborsRegressor(n_neighbors=i)
    knn.fit(x_train,y_train)
    pred_i = knn.predict(x_test)
    error_rate.append(r2_score(y_test,pred_i))
```

```python
print(np.max(error_rate))
error_rate.index(np.max(error_rate))
```

## Plotting actual values and predicted values

```python
plt.plot(data["Date"], df_new["Close"], '.',
        data["Date"], knn.predict(x), '-')
print("EPICO")
```
